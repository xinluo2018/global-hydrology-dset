```python
import cartopy.feature as cfeature
from utils.utils import add_basemap
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
path_ear5 = 'data/climate/era5_monthly_means_2020_01.nc'
proj = ccrs.PlateCarree()
```

```python
era5_xr =  xr.open_dataset(path_ear5)
era5_xr
```

```python
fig = plt.figure(figsize=(9,4))
ax = fig.add_subplot(projection=proj)
add_basemap(ax=ax)
im = ax.imshow(era5_xr['t2m'][0]-273.15, cmap='RdBu_r', extent=[-180, 180, -90, 90])
fig.colorbar(im, pad=0.1)
ax.set_title('Tempreture (C)')
```

```python
fig = plt.figure(figsize=(9,4))
ax = fig.add_subplot(projection=proj)
add_basemap(ax=ax, extent=[80, 105, 25, 40])
im = ax.imshow(era5_xr['t2m'][0]-273.15, cmap='RdBu_r', extent=[-180, 180, -90, 90])
fig.colorbar(im, pad=0.1)
ax.set_title('Tempreture (C)')
```

```python
fig = plt.figure(figsize=(9,4))
ax = fig.add_subplot(projection=proj)
add_basemap(ax=ax)
im = ax.imshow(era5_xr['tp'][0]*1000, cmap='turbo', vmax=30, vmin=0, extent=[-180, 180, -90, 90])
cbar = fig.colorbar(im, pad=0.1)
ax.set_title('Precipitation (mm)')
```
